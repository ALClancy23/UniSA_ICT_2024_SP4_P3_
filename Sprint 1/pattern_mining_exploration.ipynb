Import necessary libraries

```python
# Library for reading the csv
import pandas as pd
# Library for exploratory anslysis
import matplotlib.pyplot as plt
```

Import the chess dataset, extract the headers, view the headers and inspect the first 5 rows and get the summary statistics.

```python
# Load the dataset. Make sure to change the filepath.
file_path = 'C:\\Users\\dak_d\\Desktop\\UNI SA\\Capstone 1\\200k_blitz_rapid_classical_bullet.csv'

# Load the dataset into pandas.
data = pd.read_csv(file_path)

# Display the header names.
print(f"Header names: {data.columns}")

# Inspect the first 5 rows.
print(f"First 5 rows: \n{data.head()}")

# Display the summary statistics.
print(f"Summary statistics: \n{data.describe()}")
```

Let's do some initial investigation on openings.

```python
# Frequency Analysis: Count the frequency of different openings.
opening_counts = data['Opening'].value_counts().head(10)
print(f"Top 10 Openings:\n{opening_counts}")

# Visualise the top 10 openings.
plt.figure(figsize=(12, 6))
opening_counts.plot(kind='bar')
plt.title('Top 10 Most Frequent Openings')
plt.xlabel('Opening')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

Let's analyse the opening moves.

```python
# Analyse the top 6 opening moves for turn 1 and turn 2 only.
top_6_moves_turn_1 = data['Move_ply_1'].value_counts().head(6)
top_6_moves_turn_2 = data['Move_ply_2'].value_counts().head(6)

# Plot the top 6 opening moves for turn 1.
plt.figure(figsize=(12, 6))
top_6_moves_turn_1.plot(kind='bar', color='blue')
plt.title('Top 6 Opening Moves - Turn 1')
plt.xlabel('Move')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Plot the top 6 opening moves for turn 2.
plt.figure(figsize=(12, 6))
top_6_moves_turn_2.plot(kind='bar', color='green')
plt.title('Top 6 Opening Moves - Turn 2')
plt.xlabel('Move')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
